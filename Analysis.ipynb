#### Task#1: Merging the 12 months of sales data into a single CSV file

```python
import os
import pandas as pd

# Filter only CSV files to avoid permission errors
files = [file for file in os.listdir('./Sales_Data') if file.endswith('.csv')]

# Create an empty DataFrame to hold all the data
all_months_data = pd.DataFrame()

# Loop through each CSV file and concatenate the data
for file in files:
    df = pd.read_csv(f"./Sales_Data/{file}")
    all_months_data = pd.concat([all_months_data, df], ignore_index=True)

# Save the combined data to a new CSV file
all_months_data.to_csv("all_data.csv", index=False)
```

#### Read in updated Dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

#### Clean up the data

#### Drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

#### Columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float

all_data.info()
```

#### Augument data with additional columns 

#### Task 2:Add Month Column

```python
# Remove rows where 'Order Date' is not a proper date
all_data = all_data[all_data['Order Date'].str[0:2].str.isnumeric()]

all_data['Month'] = all_data['Order Date'].str[0:2].astype('int32')

all_data.head()
```

#### Task 3: Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Task 4: Add a city column

```python
# Let's use .apply()
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(" ")[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

#### Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month')[['Quantity Ordered','Price Each','Sales']].sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in INR')
plt.xlabel('Month Number')
plt.show()
```

##### Question 2: What city had the highest number of sales?

```python
results = all_data.groupby('City')[['Quantity Ordered','Price Each','Month','Sales']].sum()
results
```

```python
import matplotlib.pyplot as plt

cities = [city for city , df in all_data.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('Sales in INR')
plt.xlabel('City Name')
plt.show()
```

##### Question 4: What time we display advertisments to maximize likelihood of customer's buying product ?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], errors='coerce')
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour , df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.grid()
plt.show()

# My recommendation is around 11am (11) or 7pm (19)
```

#### Question4: What products are most often sold together?



```python
#- The result is a filtered DataFrame df that contains only rows where 'Order ID' appears more than once.

df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key,value)
```
